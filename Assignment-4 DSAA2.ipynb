```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class Stack:
    def __init__(self):
        self.top = None
        self.size = 0

    def push(self, value):
        new_node = Node(value)
        if not self.top:
            self.top = new_node
        else:
            new_node.next = self.top
            self.top = new_node
        self.size += 1

    def pop(self):
        if not self.top:
            return None
        else:
            popped_value = self.top.value
            self.top = self.top.next
            self.size -= 1
            return popped_value

    def __repr__(self):
        if self.size == 0:
            return f"<Stack (0 elements): []>"
        else:
            stack_list = []
            current_node = self.top
            while current_node:
                stack_list.append(current_node.value)
                current_node = current_node.next
            return f"<Stack ({self.size} element{'s' if self.size != 1 else ''}): {stack_list}>"

# Test the implementation
mystack = Stack()
mystack.push('A')
mystack.push('B')
mystack.push('C')
print(mystack)  # Output: <Stack (3 elements): ['C', 'B', 'A']>
print(mystack.pop())  # Output: C
print(mystack)  # Output: <Stack (2 elements): ['B', 'A']>
```

```python
def check_balance(text):
    stack = Stack()
    open_brackets = "([{"
    close_brackets = ")]}"
    bracket_pairs = {"(": ")", "[": "]", "{": "}"}
    error_position = -1

    for i, char in enumerate(text):
        if char in open_brackets:
            stack.push(char)
        elif char in close_brackets:
            if stack.size == 0:
                return f"Match error at position {i}"
            top = stack.pop()
            if bracket_pairs[top] != char:
                return f"Match error at position {i}"

    if stack.size != 0:
        return f"Match error at position {len(text)}"

    return f"Ok - {len(text)//2}"

# Test the function
text = "a(b)c[d]e{f}g"
print(check_balance(text))  # Output: Ok - 3
```
